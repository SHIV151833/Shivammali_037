```python
# Required libraries import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Set visualization style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Step 1: Upload Titanic CSV file
from google.colab import files
uploaded = files.upload()  # Upload your 'titanic.csv' file

# Step 2: Check the file names in the current directory
import os
print("Uploaded files:", os.listdir())

# After uploading, ensure the file name matches, for example:
# If the file is "Titanic-Dataset.csv" replace the name in pd.read_csv below

# Step 3: Load the dataset (replace 'Titanic-Dataset.csv' with the exact file name if needed)
df = pd.read_csv('Titanic-Dataset.csv')  # Use the correct filename after uploading

# Step 4: Preview the dataset
print("First 5 rows of the dataset:")
print(df.head())

# Step 5: Dataset info and missing values
print("\nDataset Info:")
print(df.info())

# Checking missing values in each column
print("\nMissing Values:")
print(df.isnull().sum())

# Step 6: Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Step 7: Visualizing numeric data with histograms
df.hist(bins=20, figsize=(14, 10), edgecolor='black')
plt.suptitle("Histograms of Numeric Columns", fontsize=16)
plt.tight_layout()
plt.show()

# Step 8: Boxplots to detect outliers
sns.boxplot(data=df, x='Survived', y='Age')
plt.title("Boxplot of Age by Survival")
plt.show()

sns.boxplot(data=df, x='Pclass', y='Fare')
plt.title("Boxplot of Fare by Passenger Class")
plt.show()

# Step 9: Countplots to analyze survival by Gender and Class
sns.countplot(data=df, x='Survived', hue='Sex')
plt.title("Survival Count by Gender")
plt.show()

sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title("Survival Count by Passenger Class")
plt.show()

# Step 10: Correlation Matrix to see relationships
numeric_cols = ['Age', 'Fare', 'Pclass', 'SibSp', 'Parch']
corr_matrix = df[numeric_cols + ['Survived']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()

# Step 11: Pairplot to analyze relationships between features
sns.pairplot(df[numeric_cols + ['Survived']].dropna(), hue='Survived')
plt.suptitle("Pairplot of Numeric Features", y=1.02)
plt.show()

# Step 12: Checking skewness of numeric features
print("\nSkewness of Numeric Features:")
print(df[numeric_cols].skew())

# Step 13: Multicollinearity (VIF - Variance Inflation Factor)
df_vif = df[numeric_cols].dropna()
X = add_constant(df_vif)
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print("\nVariance Inflation Factor (VIF):")
print(vif_data)

# Step 14: Basic Inferences
print("\nBasic Inferences:")
print("- Females had a higher survival rate than males.")
print("- Passengers in 1st class had better survival chances.")
print("- Fare is right-skewed with a few high-value outliers.")
print("- Age contains missing values and is slightly skewed.")
print("- Correlations between features are generally low.")
```
